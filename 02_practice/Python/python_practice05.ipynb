출력결과 :
```
도서명: 해리 포터와 마법사의 돌 (J.K. 롤링) - 판타지
정가: 15000원
할인가: 13500원

도서명: 사피엔스: 인류의 대역사 (유발 하라리) - 역사

도서명: 1984 (조지 오웰) - 디스토피아 소설

할인가: 15300원

도서명: 지적 대화를 위한 넓고 얕은 지식 (채사장) - 인문학

할인가: 16150원

도서명: 어린 왕자 (앙투안 드 생텍쥐페리) - 고전 문학
정가: 12000원
할인가: 9000원
```

````python
class Book:
    def __init__(self, title, author, genre, price, discount_rate):
        """
        도서 객체 초기화
        :param title: 도서명
        :param author: 저자
        :param genre: 장르
        :param price: 정가
        :param discount_rate: 할인율 (0.0 ~ 1.0)
        """
        self.title = title
        self.author = author
        self.genre = genre
        self.price = price
        self.discount_rate = discount_rate

    def book_information(self):
        """
        도서 정보를 지정된 형식으로 출력
        할인가 계산: 정가 * (1 - 할인율)
        """
        # 할인가 계산 (정수형으로 변환)
        discounted_price = int(self.price * (1 - self.discount_rate))
        
        print(f"도서명: {self.title} ({self.author}) - {self.genre}")
        print(f"정가: {self.price}원")
        print(f"할인가: {discounted_price}원")
        print()  # 책 정보 간의 공백 라인 추가


# --- 실행 코드 ---

# 샘플 도서 5권 생성 (다양한 장르)
book1 = Book("해리 포터와 마법사의 돌", "J.K. 롤링", "판타지", 15000, 0.1) 
book2 = Book("사피엔스: 인류의 대역사", "유발 하라리", "역사", 22000, 0.2) 
book3 = Book("1984", "조지 오웰", "디스토피아 소설",  18000, 0.15) 
book4 = Book("지적 대화를 위한 넓고 얕은 지식", "채사장", "인문학", 17000, 0.05) 
book5 = Book("어린 왕자", "앙투안 드 생텍쥐페리", "고전 문학", 12000, 0.25) 

# 도서 정보 출력
book1.book_information()
book2.book_information()
book3.book_information()
book4.book_information()
book5.book_information()
````
