````python
from enum import Enum


class RestaurantType(Enum):
    KOR = "í•œ"
    CHN = "ì¤‘"
    JPN = "ì¼"


class Food(Enum):
    MEAL = "ì‹ì‚¬"
    SNACK = "ì•ˆì£¼"
    DRINK = "ìŒë£Œ"


class MenuItem:
    def __init__(self, name, price, foodType: Food, num=0):
        self.__name = name
        self.__price = price
        if isinstance(foodType, Food):
            self.__foodType = foodType
        else:
            self.__foodType = Food(foodType)
        self.__num = num

    @property
    def foodType(self):
        return self.__foodType

    @property
    def name(self):
        return self.__name

    @property
    def price(self):
        return self.__price

    @property
    def num(self):
        return self.__num

    @num.setter
    def num(self, num):
        self.__num = num


class Restaurant:
    def __init__(self, name, address, tel, restaurant_type):
        self.__name = name
        self.__address = address
        self.__tel = tel
        self.__menu: list[MenuItem] = []
        self.__restaurant_type: RestaurantType = restaurant_type

    def add_menu(self, menuitem: MenuItem):
        self.__menu.append(menuitem)

    @property
    def name(self):
        return self.__name

    @property
    def address(self):
        return self.__address

    @property
    def tel(self):
        return self.__tel

    @property
    def menu_items(self) -> tuple[MenuItem]:
        return tuple(self.__menu)

    @property
    def restaurant_type(self) -> RestaurantType:
        return self.__restaurant_type

    @restaurant_type.setter
    def restaurant_type(self, type: RestaurantType):
        self.__restaurant_type = type

    def display_info(self):
        print(
            f"""[{self.__restaurant_type.value}ì‹ë‹¹ - {self.name}]
<ì£¼ì†Œ: {self.address}>
<ì „í™”ë²ˆí˜¸: {self.tel}>
"""
        )

    def display_menu(self):
        for menu in self.menu_items:
            print(f"{menu.name} - {menu.price}ì›")
        print("-" * 70)


class KoreanRestaurant(Restaurant):
    def __init__(self, name, address, tel):
        super().__init__(name, address, tel, RestaurantType.KOR)


class JapaneseRestaurant(Restaurant):
    def __init__(self, name, address, tel):
        super().__init__(name, address, tel, RestaurantType.JPN)


class ChineseRestaurant(Restaurant):
    def __init__(self, name, address, tel):
        super().__init__(name, address, tel, RestaurantType.CHN)


class Order:
    def __init__(self, restaurant: Restaurant):
        self.__restaurant = restaurant
        self.__items: list[MenuItem] = []
        self.__total_price = 0

    def add_item(self, menu_item: MenuItem, num):
        menu_item.num = num
        self.__items.append(menu_item)
        self.__total_price += menu_item.price * num

    def display_order(self):
        print(
            f"[{self.__restaurant.restaurant_type.value}ì‹ë‹¹ - {self.__restaurant.name}]\n"
        )
        for item in self.__items:
            print(
                f"{item.name} - {item.price}ì› * {item.num}ê°œ {item.price * item.num}ì›"
            )
        print(f"\nì´ê²°ì œê¸ˆì•¡: {self.__total_price}ì›")
````
