### Î¨∏Ï†ú 1. ÏßÅÏ†ë Í∞ÄÏ§ëÏπò/Ï†àÌé∏ÏùÑ ÏÑ†Ïñ∏ÌïòÍ≥† ÏàúÏ†ÑÌåå Í≥ÑÏÇ∞

Îã§Ïùå Ï°∞Í±¥Ïóê ÎßûÍ≤å PyTorch ÌÖêÏÑúÎ•º ÏÇ¨Ïö©ÌïòÏó¨ forward Í≥ÑÏÇ∞ÏùÑ ÏàòÎèôÏúºÎ°ú Íµ¨ÌòÑÌïòÏÑ∏Ïöî.

**üìå ÏöîÍµ¨ÏÇ¨Ìï≠**

1. ÏûÖÎ†• `X`Îäî `torch.randn(6, 10)` ÏúºÎ°ú ÏÉùÏÑ±ÌïúÎã§.
2. `W1`, `b1`, `W2`, `b2`ÎèÑ torchÏùò `randn`ÏùÑ Ïù¥Ïö©Ìï¥ ÏßÅÏ†ë Ï†ïÏùòÌïúÎã§.
3. ÏùÄÎãâÏ∏µÏóêÏÑúÎäî `tanh`, Ï∂úÎ†•Ï∏µÏóêÏÑúÎäî `softmax`Î•º ÏÇ¨Ïö©ÌïúÎã§.
4. ÏµúÏ¢Ö Í≤∞Í≥ºÎäî `(6, 4)` ÌÅ¨Í∏∞Ïùò ÌôïÎ•† Î≤°ÌÑ∞Ïù¥Ïñ¥Ïïº ÌïúÎã§.

```python
import torch
import torch.nn.functional as F

# 1Ô∏è‚É£ ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞
X = torch.randn(6, 10)   # (ÏÉòÌîå 6, ÌäπÏÑ± 10)

# 2Ô∏è‚É£ Í∞ÄÏ§ëÏπòÏôÄ Ìé∏Ìñ• ÏßÅÏ†ë ÏÑ†Ïñ∏
W1 = torch.randn(10, 8)
b1 = torch.randn(8)

W2 = torch.randn(8, 4)
b2 = torch.randn(4)

# 3Ô∏è‚É£ ÏàúÏ†ÑÌåå Í≥ÑÏÇ∞
z1 = X @ W1 + b1         # (6, 8)
a1 = torch.tanh(z1)     # tanh ÌôúÏÑ±Ìôî

z2 = a1 @ W2 + b2       # (6, 4)
output = F.softmax(z2, dim=1)  # softmax

# 4Ô∏è‚É£ Í≤∞Í≥º ÌôïÏù∏
print(output.shape)
print(output)
```

### üí° Ï°∞Í±¥

* Ï∂úÎ†•Ï∏µ: ÏûÖÎ†• 8 ‚Üí Ï∂úÎ†• 4 (ÌôúÏÑ±Ìôî Ìï®Ïàò: `softmax`)

### Î¨∏Ï†ú 2. ÏúÑ ÎÑ§Ìä∏ÏõåÌÅ¨Î•º PyTorch Î™®Îç∏ ÌÅ¥ÎûòÏä§Î°ú ÏûëÏÑ±

Îã§Ïùå Ï°∞Í±¥Ïóê ÎßûÎäî `nn.Module` ÌÅ¥ÎûòÏä§Î•º ÏûëÏÑ±ÌïòÏÑ∏Ïöî.

üìå ÏöîÍµ¨ÏÇ¨Ìï≠

1. `__init__`Ïóê `nn.Linear(10, 8)`Í≥º `nn.Linear(8, 4)`Î•º Ï†ïÏùòÌïúÎã§.
2. `forward`ÏóêÏÑúÎäî `tanh` ‚Üí `softmax` ÏàúÏúºÎ°ú ÌôúÏÑ±Ìôî Ìï®ÏàòÎ•º Ï†ÅÏö©ÌïúÎã§.
3. ÏûÑÏùòÏùò ÏûÖÎ†• `X = torch.randn(6, 10)`ÏùÑ Î™®Îç∏Ïóê ÎÑ£Ïñ¥ Ï∂úÎ†• Í≤∞Í≥ºÎ•º ÌôïÏù∏ÌïúÎã§.


```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class SimpleNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(10, 8)
        self.fc2 = nn.Linear(8, 4)

    def forward(self, x):
        x = torch.tanh(self.fc1(x))
        x = F.softmax(self.fc2(x), dim=1)
        return x

# Î™®Îç∏ ÏÉùÏÑ±
model = SimpleNN()

# ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞
X = torch.randn(6, 10)

# ÏàúÏ†ÑÌåå
output = model(X)

print(output.shape)
print(output)
```

* Ï∂úÎ†•Ï∏µ: ÏûÖÎ†• 8 ‚Üí Ï∂úÎ†• 4 (ÌôúÏÑ±Ìôî Ìï®Ïàò: `softmax`)

```python
# Ï≤´ Î≤àÏß∏ Ï∏µ
print("FC1 Weight:\n", model.fc1.weight)
print("FC1 Bias:\n", model.fc1.bias)

# Îëê Î≤àÏß∏ Ï∏µ
print("FC2 Weight:\n", model.fc2.weight)
print("FC2 Bias:\n", model.fc2.bias)
```
