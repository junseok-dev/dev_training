## Level 1: 데이터 로드

```
Online Retail 데이터셋을 로드한 후:
1) 데이터 형태(shape) 출력
2) 결측치 개수 확인
3) 고객 수, 상품 수 계산
```


````python
# 데이터 로드
retail_df = pd.read_excel('https://d.pr/rsMK94+')
````

````python
# 1) 데이터 형태(shape) 출력
print(retail_df.shape)

# 2) 결측치 개수 확인
print(retail_df.info())

# 3) 고객 수, 상품 수 계산
print(retail_df['CustomerID'].value_counts().sum())
print(retail_df['StockCode'].value_counts().sum())
````


## Level 2: 데이터 전처리

**문제 2-1: 데이터 정제**
```
다음 조건에 따라 데이터를 정제하세요:
1) 결측치 있는 행 제거
2) Quantity <= 0 인 행 제거 (반품)
3) CustomerID를 정수로 변환
```


````python
# 1)
retail_df = retail_df[~retail_df['CustomerID'].isnull()]

# 2)
retail_df = retail_df[~(retail_df['Quantity'] <= 0)]

# 3)
retail_df['CustomerID'] = retail_df['CustomerID'].astype(int)
````


**문제 2-2: 구매 통계**



````python
# 1)
display(retail_df.groupby('CustomerID')['Quantity'].mean())

# 2)
display(retail_df.groupby('StockCode')['StockCode'].count().mean())

# 3)
quantity_ser = retail_df.groupby('StockCode')['Quantity'].sum()
quantity_ser.sort_values(ascending=False).head(3)
````


## Level 3: 사용자-상품 행렬 생성

**문제 3-1: 기본 행렬 생성**
```
사용자-상품 행렬을 생성하세요:
- 행: CustomerID
- 열: StockCode
- 값: 구매 Quantity의 합
→ 행렬 크기 출력
```


````python
pivot_retail = retail_df.pivot_table(columns='CustomerID', index='StockCode', values='Quantity', fill_value=0)
print(pivot_retail.shape)
````


**문제 3-2: 행렬 특성 분석**
```
생성한 행렬에서:
1) 희소성(Sparsity) 계산
   [0이 아닌 원소 / 전체 원소]
2) 각 고객의 구매 상품 개수 분포
3) 각 상품의 구매 고객 개수 분포
```


````python
# 1)
total_element = pivot_retail.size
nonzero_element = np.count_nonzero(pivot_retail.values)
sparsity = nonzero_element/total_element
print(f'sparsity : {sparsity}')

# 2)
customer_product_counts = (pivot_retail > 0).sum(axis=0)
display(customer_product_counts.head())

# 3)
product_customer_counts = (pivot_retail > 0).sum(axis=1)
display(product_customer_counts.head())
````


## Level 4: 코사인 유사도 계산

**문제 4-1: 상품 간 유사도**
```
cosine_similarity()를 사용하여 상품 간 유사도 행렬을 생성하세요.

→ 첫 번째 상품과 유사한 상품 TOP 5 출력
  (자신과의 유사도 1.0 제외)
```


````python
from sklearn.metrics.pairwise import cosine_similarity

# 1) 유사도 행렬
item_sim = cosine_similarity(pivot_retail)

# 2) 첫 번째 상품의 유사도
first_item_sim = item_sim[0]

# 3) 자기 자신 제외한 TOP 5
sorted_idx = np.argsort(first_item_sim)[::-1]
top5_idx = sorted_idx[1:6]

# 4) 상품 코드 출력
top5_items = pivot_retail.index[top5_idx]
list(zip(top5_items, first_item_sim[top5_idx]))
````


**문제 4-2: 유사도 분석**
```
생성한 유사도 행렬에서:
1) 전체 유사도의 통계
   - 평균, 최소값, 최대값, 중앙값
2) 유사도 >= 0.7 인 상품 쌍의 개수
3) 가장 유사한 상품 쌍 3개
```


````python
item_sim_df = pd.DataFrame(item_sim)
# 1)
display(item_sim_df.describe())

# 2)
mask = (item_sim >= 0.7) & np.triu(np.ones_like(item_sim), k=1).astype(bool)
pair_count = mask.sum()
print(pair_count)

# 3)
# 상삼각 mask (자기 자신 제외, 중복 제거)
mask = np.triu(np.ones(item_sim.shape), k=1).astype(bool)

# 유사도 값 중에서 상삼각 부분만 추출
upper_vals = item_sim[mask]

# 상삼각에서 가장 유사한 3개 값의 인덱스 찾기
top3_idx = np.argsort(upper_vals)[-3:]  # 큰 값 TOP 3

# 전체 행렬 좌표로 변환
coords = np.column_stack(np.where(mask))
top3_pairs = coords[top3_idx]

# 결과 정리
results = []
for i, j in top3_pairs:
    item1 = pivot_retail.index[i]
    item2 = pivot_retail.index[j]
    sim = item_sim[i, j]
    results.append((item1, item2, sim))

results
````


## Level 5: 추천 함수 구현 (상품 기반)

**문제 5-1: 단순 추천 함수**
```
함수를 완성하세요:

def recommend_items(product_id, n=5):
    """주어진 상품과 유사한 상품 추천"""
    1) product_id가 존재하는지 확인
    2) 유사도를 내림차순으로 정렬
    3) 자신 제외하고 상위 n개 반환
```


````python
def recommend_items(product_id, n=5):
    """주어진 상품과 유사한 상품 추천"""
    
    # 1) product_id 존재 여부 확인
    if product_id not in pivot_retail.index:
        raise ValueError(f"{product_id} 는 존재하지 않는 상품입니다.")
    
    # 2) 해당 상품의 인덱스 얻기
    idx = pivot_retail.index.get_loc(product_id)
    
    # 3) 해당 상품의 유사도 벡터 가져오기
    sim_vector = item_sim[idx]
    
    # 4) 유사도 내림차순 정렬 (가장 유사한 상품이 앞에 오도록)
    sorted_idx = np.argsort(sim_vector)[::-1]
    
    # 5) 자기 자신(유사도 1.0) 제외 후 상위 n개 선택
    recommended_idx = sorted_idx[1:n+1]     # [1:] ← 자기 자신 제거
    
    # 6) 실제 상품 ID로 변환
    recommended_items = pivot_retail.index[recommended_idx]
    
    # 7) 상품 ID와 유사도 함께 출력용 묶음 생성
    return list(zip(recommended_items, sim_vector[recommended_idx]))

print(recommend_items('84535A'))
print(recommend_items('90177E'))
print(recommend_items('79323P'))
print(recommend_items('90141E'))
print(recommend_items('90059A'))
````


**문제 5-2: 함수 테스트**
```
구현한 함수를 테스트하세요:
1) 5개의 다른 상품에 대해 추천 실행
2) 각 결과에서 유사도 점수 확인
3) 논리적으로 타당한지 검증
```



````python
test_products = pivot_retail.index[:5]   # 앞에서 5개 상품 선택

for pid in test_products:
    print(f"\n=== 상품 {pid} 추천 결과 ===")
    recs = recommend_items(pid, n=5)
    for item, sim in recs:
        print(f"  → {item} (유사도: {sim:.4f})")
````


## Level 6: 추천 함수 구현 (고객 기반)

**문제 6-1: 고객 맞춤 추천 함수**
```
함수를 완성하세요:

def recommend_for_customer(customer_id, n=5):
    """고객이 구매한 상품과 유사한 상품 추천"""
    1) 고객의 구매 상품 목록 추출
    2) 각 구매 상품과 유사한 상품 찾기
    3) 이미 구매한 상품 제외
    4) 유사도 점수 누적하여 상위 n개 반환
```

````python
def recommend_for_customer(customer_id, n=5):
    """고객이 구매한 상품과 유사한 상품 추천"""
    
    # 1) 고객 존재 여부 확인
    if customer_id not in pivot_retail.columns:
        raise ValueError(f"{customer_id} 는 존재하지 않는 고객입니다.")
    
    # 2) 고객이 구매한 상품 목록 추출
    # pivot_retail에서 해당 고객 column에서 값 > 0인 상품들
    purchased_items = pivot_retail[pivot_retail[customer_id] > 0].index.tolist()
    
    if len(purchased_items) == 0:
        return f"{customer_id} 고객은 구매 이력이 없습니다."
    
    # 3) 추천 점수를 누적할 dictionary
    scores = {}
    
    for item in purchased_items:
        # item의 index 찾기
        idx = pivot_retail.index.get_loc(item)
        
        # item과 모든 상품 간 유사도 벡터
        sim_vec = item_sim[idx]
        
        # 유사도 누적
        for i, sim in enumerate(sim_vec):
            target_item = pivot_retail.index[i]
            
            # 본인이 구매한 상품은 제외
            if target_item in purchased_items:
                continue
            
            # 유사도 누적 점수 계산
            scores[target_item] = scores.get(target_item, 0) + sim
    
    # 4) 유사도 점수로 상품 정렬 (내림차순)
    sorted_items = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    
    # 5) 상위 n개 반환
    return sorted_items[:n]
"""
12346
12347
12348
12349
12350
"""
print(recommend_for_customer(12346))
print(recommend_for_customer(12347))
print(recommend_for_customer(12348))
print(recommend_for_customer(12349))
print(recommend_for_customer(12350))
````
