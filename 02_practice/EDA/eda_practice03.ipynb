**문제 1: DataFrame객체 생성**

위의 데이터를 사용해 DataFrame객체를 생성하세요.

```python
import numpy as np
import pandas as pd

data = {
    'Name':['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age':[25, 30, 35, 40, 28],
    'City':['New York', 'Los Angeles', 'Chicago', 'Houston', 'San Francisco'],
    'Salary':[70000, 80000, 120000, 90000, 75000]
}

df = pd.DataFrame(data)
print(df.info())
df
```

# 문제 2: 열과 행 선택

```python
# 1. Name과 City 열 선택
print(df[['Name', 'City']])

# 2. 나이가 30 이상인 행 선택
print(df[df['Age'] >= 30])
```

# 문제 3: 행과 열 추가

```python
# 1. Country 열 추가 (모두 'USA')
df['Country'] = 'USA'
print(df)

# 2. 새로운 행 추가 (Frank, 33, Seattle, 85000)
new_row = {'Name': 'Frank', 'Age': 33, 'City': 'Seattle', 'Salary': 85000, 'Country': 'USA'}
df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
print(df)
```

# 문제 4: 데이터 삭제 

```python
# 1. Salary 열 삭제
df = df.drop(columns='Salary')
print(df)

# 2. 'David'가 있는 행 삭제
df = df[df['Name'] != 'David']
print(df)
```

## 문제 5: 데이터 수정

Age가 35 이상인 사람들의 Salary를 5% 증가시키세요.
'Alice'의 City를 'Boston'으로 수정하세요.

```python
# 데이터 다시 로드 후 실행 
# # df.loc[df['Age'] >= 35, 'Salary'] = df.loc[df['Age'] >= 35, 'Salary'] * 1.05
df.loc[df['Age'] >= 35, 'Salary'] *= 1.05
df
```

```python
df.loc[df['Name'] == 'Alice', 'City'] = 'Boston'
df
```

```python
# 검색
df[df['Name'] == 'Alice'] # []표현식
df.loc[df['Name'] == 'Alice', 'Age':'City'] # loc[] 인덱스/컬럼 라벨
```

```python
import pandas as pd

# 1. 데이터 로드 (파일 경로에 맞게 수정하세요)
df = pd.read_csv('data/titanic.csv') 

# (분석에 사용된 데이터 예시)
# ... (데이터 로드 부분 생략) ...

# 2. 성별(Sex)과 선실 등급(Pclass)별 생존율(Survived) 집계
# pivot_table을 사용하면 표 형태로 보기 좋게 정리됩니다.
survival_pivot = df.pivot_table(index='Sex', columns='Pclass', values='Survived', aggfunc='mean')

print("성별 및 클래스별 생존율:")
print(survival_pivot)
```
