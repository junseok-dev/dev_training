```python
import numpy as np

scores = np.array([
    [87, 90, 85],  # 철수
    [92, 88, 79],  # 영희
    [76, 95, 90],  # 민수
    [88, 84, 92]   # 수진
])

scores
```

```python
scores[:, 1]
```

```python
scores[[0, 3], 2]
```

# NumPy 실습문제 3: 조건 인덱싱 + 통계 연산

## 점수 데이터 생성

```python
np.random.seed(0)
scores = np.random.randint(0, 101, size=100)
```

## 1. 90점 이상 학생 수

```python
np.sum(scores >= 90)
```

## 2. 50점 미만 점수 평균

```python
low_scores = scores[scores < 50]
np.mean(low_scores)
```

## 3. 전체 평균 / 최대 / 최소

```python
np.mean(scores), np.max(scores), np.min(scores)
```

# NumPy 실습문제 4: 다차원 배열 연산 및 axis

## 데이터 생성

```python
data = np.array([
    [23, 25, 22],
    [24, 26, 23],
    [22, 24, 21],
    [25, 27, 24],
    [23, 25, 22]
])
```

## 1. 센서별 평균 (axis=0)

```python
np.mean(data, axis=0)
```

## 2. 날짜별 합계 (axis=1)

```python
np.sum(data, axis=1)
```

## 3. 전체 최소 / 최대

```python
np.min(data), np.max(data)
```

# NumPy 실습문제 5: 행렬 곱과 브로드캐스팅

## 행렬 정의

```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[1, 0],
              [0, 1],
              [1, 1]])
```

## 1. 행렬 곱

```python
C = A @ B
C
```

## 2. 브로드캐스팅으로 값 더하기

```python
C + np.array([10, 100])
```

## 3. 최종 결과 shape

```python
C.shape
```
