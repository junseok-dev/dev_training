### **학생별 과목 성적 데이터**
   - 데이터: 한 학교의 10명의 학생들이 수학, 영어, 과학 시험에서 받은 점수 (3과목)

**문제:**
1. **학생들의 수학 성적**을 이용해 **평균**, **중앙값**, **최빈값**을 구하시오.
2. **수학 성적**의 **분산**과 **표준편차**를 구하여, 학생들의 성적이 얼마나 퍼져 있는지 확인하시오.
3. **영어 성적**의 **평균**과 **중앙값**을 구하고, 이 값들을 비교하여 성적의 분포를 해석하시오.
4. **과학 성적**의 **최빈값**을 구하고, 만약 여러 개의 최빈값이 존재한다면 모두 출력하시오.
5. 각 학생의 **전체 평균 성적**(수학, 영어, 과학)을 계산한 후, 이 평균값에 대한 **분산**과 **표준편차**를 구하시오.

**데이터:**

| 학생 | 수학 | 영어 | 과학 |
| ---- | ---- | ---- | ---- |
| 1    | 88   | 90   | 85   |
| 2    | 92   | 85   | 88   |
| 3    | 76   | 80   | 82   |
| 4    | 85   | 87   | 89   |
| 5    | 95   | 92   | 94   |
| 6    | 92   | 89   | 86   |
| 7    | 88   | 83   | 91   |
| 8    | 90   | 91   | 88   |
| 9    | 85   | 86   | 84   |
| 10   | 88   | 88   | 87   |

```python
# 학생별 과목 성적 데이터
scores_math = [88, 92, 76, 85, 95, 92, 88, 90, 85, 88]
scores_english = [90, 85, 80, 87, 92, 89, 83, 91, 86, 88]
scores_science = [85, 88, 82, 89, 94, 86, 91, 88, 84, 87]
```

# 1. 수학 성적 - 평균, 중앙값, 최빈값

````python
import numpy as np
from scipy import stats

scores_math = [88, 92, 76, 85, 95, 92, 88, 90, 85, 88]

# 평균
mean_math = np.mean(scores_math)
# 중앙값
median_math = np.median(scores_math)
# 최빈값 (여러 개 존재할 수 있음)
mode_math = stats.mode(scores_math, keepdims=True).mode

print("수학 성적 - 평균:", mean_math)
print("수학 성적 - 중앙값:", median_math)
print("수학 성적 - 최빈값:", mode_math)
````

# 2. 수학 성적 - 분산과 표준편차

````python
# 분산
var_math = np.var(scores_math, ddof=0)  # 모집단 기준
# 표준편차
std_math = np.std(scores_math, ddof=0)

print("수학 성적 - 분산:", var_math)
print("수학 성적 - 표준편차:", std_math)
````

# 3. 영어 성적 - 평균과 중앙값 비교

````python
scores_english = [90, 85, 80, 87, 92, 89, 83, 91, 86, 88]

mean_eng = np.mean(scores_english)
median_eng = np.median(scores_english)

print("영어 성적 - 평균:", mean_eng)
print("영어 성적 - 중앙값:", median_eng)
````

# 4. 과학 성적 - 최빈값 (중복 가능)

````python
scores_science = [85, 88, 82, 89, 94, 86, 91, 88, 84, 87]

mode_sci = stats.mode(scores_science, keepdims=True).mode

print("과학 성적 - 최빈값:", mode_sci)
````

# 5. 학생별 전체 평균 성적 및 분산, 표준편차

````python
# 학생별 전체 평균
scores_all = np.array([scores_math, scores_english, scores_science]).T
avg_per_student = np.mean(scores_all, axis=1)

# 전체 평균의 분산과 표준편차
var_avg = np.var(avg_per_student, ddof=0)
std_avg = np.std(avg_per_student, ddof=0)

print("학생별 전체 평균 성적:", avg_per_student)
print("전체 평균의 분산:", var_avg)
print("전체 평균의 표준편차:", std_avg)
````
