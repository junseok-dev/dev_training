50명의 승객과 매칭기회가 있을 때, 총 탑승 승객 수를 구하는 프로그램을 작성하시오.

- 조건1: 승객별 운행 소요 시간은 `5분 ~ 50분` 사이의 난수로 정해집니다.
- 조건2: 당신은 소요 시간 `5분 ~ 15분` 사이의 승객만 매칭해야 합니다.

**(출력문 예제)**
```
[0] 1번째 손님 (소요시간 : 15분)
[] 2번째 손님 (소요시간 : 50분)
[0] 3번째 손님 (소요시간 : 5분)
...
[] 50번째 손님 (소요시간 : 16분)

총 탑승 승객: 2 분
```

````python
# 나혼자 코딩 인강 정답 코드 

from random import *

cnt = 0 # 총 탑승 승객 수

for i in range(1, 51): # 1 ~ 50 이라는 수 (승객)
    time = randrange(5, 51) # 5분 ~ 50분 소요 시간
    if 5 <= time <= 15: # 5분 ~ 15분 이내의 손님, 탑승 승객 수 증가 처리 
        print("[0] {0}번째 손님 (소요시간 : {1}분)".format(i, time))
    else: # 매칭 실패한 경우
        print("[ ] {0}번째 손님 (소요시간 : {1}분)".format(i, time))

print("총 탑승 승객 : 2 분".format(cnt))
````

````python
# ai 로 만든 코드 

from random import *

cnt = 0 # 총 탑승 승객 수 (0으로 초기화)

for i in range(1, 51): # 1 ~ 50번째 승객
    time = randrange(5, 51) # 5분 ~ 50분 소요 시간
    
    if 5 <= time <= 15: # [조건] 5분 ~ 15분 사이인 경우만!
        print("[0] {0}번째 손님 (소요시간 : {1}분)".format(i, time))
        cnt += 1 # ⭐ 중요: 조건에 맞을 때만 카운트를 1 증가시킴
    else: 
        print("[ ] {0}번째 손님 (소요시간 : {1}분)".format(i, time))

# 모든 반복이 끝난 후 최종 합계 출력
print("\n총 탑승 승객 : {0} 명".format(cnt))
````
