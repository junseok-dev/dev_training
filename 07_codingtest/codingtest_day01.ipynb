# 📝 문제: 마지막 두 원소

### **1. 문제 설명**
정수 리스트 `num_list`가 주어질 때, 다음 조건에 따라 값을 추가하여 return하도록 solution 함수를 완성해주세요.

1.  **마지막 원소**가 **그전 원소**보다 **크면**: 
    - `마지막 원소 - 그전 원소` 값을 추가
2.  **마지막 원소**가 **그전 원소**보다 **크지 않다면(작거나 같으면)**: 
    - `마지막 원소 * 2` 값을 추가

---

### **2. 제한사항**
- 2 ≤ `num_list`의 길이 ≤ 10
- 1 ≤ `num_list`의 원소 ≤ 9

---

### **3. 입출력 예**

| num_list | result |
| :--- | :--- |
| [2, 1, 6] | [2, 1, 6, 5] |
| [5, 2, 1, 7, 5] | [5, 2, 1, 7, 5, 10] |

---

### **4. 입출력 예 설명**
- **예제 #1**: 마지막 원소(6)가 이전 원소(1)보다 크므로 `6 - 1 = 5`를 추가합니다.
- **예제 #2**: 마지막 원소(5)가 이전 원소(7)보다 크지 않으므로 `5 * 2 = 10`를 추가합니다.

---

````python
# 문제 1번 정답 확인 코드

# 프로그래머스 정답
def solution(num_list):
    # 기본 틀의 answer에 원본 리스트를 복사해둡니다.
    answer = num_list[:]
    
    # 마지막 원소(last)와 그전 원소(prev)를 변수에 담습니다.
    last = num_list[-1]
    prev = num_list[-2]
    
    # 1. 마지막 원소가 그전 원소보다 크면
    if last > prev:
        answer.append(last - prev)
    # 2. 마지막 원소가 그전 원소보다 크지 않다면 (작거나 같으면)
    else:
        answer.append(last * 2)
        
    return answer


# 테스트 케이스 실행
test_cases = [
    {"input": [2, 1, 6], "expected": [2, 1, 6, 5]},
    {"input": [5, 2, 1, 7, 5], "expected": [5, 2, 1, 7, 5, 10]}
]

for i, test in enumerate(test_cases):
    result = solution(test["input"])
    print(f"입출력 예 #{i+1}")
    print(f"  - 입력값: {test['input']}")
    print(f"  - 결과값: {result}")
    print(f"  - 정답여부: {'✅ 통과' if result == test['expected'] else '❌ 실패'}")
    print("-" * 35)

# 방법 1
# def solution(num_list):
#     last = num_list[-1]
#     prev = num_list[-2]

#     if last > prev:
#         num_list.append(last-prev)
#     else:
#         num_list.append(last * 2)
    
#     return num_list

# 2. 방법 1 함수 실행 및 결과 확인 (이 부분이 있어야 눈으로 볼 수 있어요)
# [2, 1, 6]을 넣어서 실행한 결과를 result 변수에 담습니다.
# result = solution([2, 1, 6]) 

# # 결과를 화면에 출력합니다.
# print(result)

# result2 = solution([5,2,1,7,5])
# print(result2)
````

# 문제 2번 

# 📝 문제: 콜라츠 수열 만들기

### **1. 문제 설명**
모든 자연수 $x$에 대해서 현재 값이 $x$이면 다음과 같은 규칙을 반복합니다.
- $x$가 **짝수**일 때: $x / 2$
- $x$가 **홀수**일 때: $3 * x + 1$

이 계산을 반복하면 언젠가는 반드시 $x$가 1이 된다는 것이 콜라츠 문제이며, 이 과정에서 거쳐간 모든 수를 기록한 수열을 **콜라츠 수열**이라고 부릅니다.

임의의 1,000보다 작거나 같은 양의 정수 `n`이 주어질 때, 초기값이 `n`인 콜라츠 수열을 return 하는 solution 함수를 완성해 주세요.

---

### **2. 제한사항**
- 1 ≤ `n` ≤ 1,000

---

### **3. 입출력 예**

| n | result |
| :--- | :--- |
| 10 | [10, 5, 16, 8, 4, 2, 1] |

**입출력 예 설명**
- 10(짝수) → 5(홀수) → 16(짝수) → 8(짝수) → 4(짝수) → 2(짝수) → 1
- 위 순서대로 연산한 결과를 담은 리스트 `[10, 5, 16, 8, 4, 2, 1]`을 반환합니다.

---

````python
# 문제 2번 정답 확인 코드

# 1. 솔루션 함수 (작성한 로직)
def solution(n):
    answer = [n]  # 초기값 n을 리스트에 먼저 넣습니다.
    
    while n != 1:  # n이 1이 아닐 때까지 반복
        if n % 2 == 0:      # 짝수라면
            n = n // 2      # 2로 나눔 (//는 정수 나누기)
        else:               # 홀수라면
            n = 3 * n + 1   # 3을 곱하고 1을 더함
            
        answer.append(n)    # 계산된 n을 리스트에 추가
        
    return answer

# ---------------------------------------------------------

# 2. 정답 검증 로직
# 문제에 나온 예시: 10을 넣으면 [10, 5, 16, 8, 4, 2, 1]이 나와야 함
input_n = 10
expected_result = [10, 5, 16, 8, 4, 2, 1]

# 함수 실행
my_result = solution(input_n)

# 3. 결과 비교 및 출력
print(f"입력값 : {input_n}")
print(f"기댓값 : {expected_result}") # 문제에서 알려준 정답
print(f"내결과 : {my_result}")      # 내 함수가 계산한 값

print("-" * 30) # 구분선

if my_result == expected_result:
    print("✅ 테스트 통과! 정답과 완벽하게 일치합니다.")
else:
    print("❌ 테스트 실패... 정답과 다릅니다.")
````

````python
# 문제 2번 (복습 필요)

def solution(n):
    # 초기값 n을 리스트에 넣음 
    answer = [n]
    # n이 1이 아닐 때 까지 반복
    while n != 1:
        # 짝수라면
        if n % 2 == 0:
           # n을 2로 나눔
           n = n // 2
        # 홀수라면
        else:
            # n을 3을 곱하고 1을 더함
            n = 3 * n + 1
        # 계산된 n을 리스트에 추가 
        answer.append(n)

    return answer

# 10은 짝수 -> 2로 나눔 -> 5
# 5는 홀수 3을 곱하고 1을 더함 -> 16
# 16은 짝수 -> 2로 나눔 -> 8
# 8은 짝수 -> 2로 나눔 -> 4
# 4는 짝수 -> 2로 나눔 -> 2
# 2는 짝수 -> 2로 나눔 -> 1
# 10 -> [10, 5, 16, 8, 4, 2, 1]
````

# 문제 10번

# [프로그래머스] 주사위 게임 1

### 📝 문제 설명
1부터 6까지 숫자가 적힌 주사위가 두 개 있습니다. 두 주사위를 굴렸을 때 나온 숫자를 각각 `a`, `b`라고 했을 때 얻는 점수는 다음과 같습니다.

1.  **`a`와 `b`가 모두 홀수**라면: $a^2 + b^2$ 점
2.  **`a`와 `b` 중 하나만 홀수**라면: $2 \times (a + b)$ 점
3.  **`a`와 `b` 모두 홀수가 아니라면(모두 짝수)**: $|a - b|$ 점

두 정수 `a`와 `b`가 매개변수로 주어질 때, 얻는 점수를 return 하는 `solution` 함수를 작성해 주세요.

---

### 🚫 제한사항
* `a`와 `b`는 1 이상 6 이하의 정수입니다.

---
### 📥 입출력 예
| a | b | result |
| :--- | :--- | :--- |
| 3 | 5 | 34 |
| 6 | 1 | 14 |
| 2 | 4 | 2 |

---

### 💡 입출력 예 설명

**입출력 예 #1**
* 3과 5는 모두 홀수입니다. $3^2 + 5^2 = 9 + 25 = 34$점을 얻습니다.

**입출력 예 #2**
* 6은 짝수, 1은 홀수입니다. 하나만 홀수이므로 $2 \times (6 + 1) = 14$점을 얻습니다.

**입출력 예 #3**
* 2와 4는 모두 짝수(홀수가 아님)입니다. $|2 - 4| = |-2| = 2$점을 얻습니다.

---

### 💻 정답 코드 (Python)

```python
def solution(a, b):
    # 홀수인지 여부를 먼저 판단합니다.
    is_a_odd = a % 2 == 1
    is_b_odd = b % 2 == 1
    
    # 1. 둘 다 홀수인 경우
    if is_a_odd and is_b_odd:
        return a**2 + b**2
    
    # 2. 둘 중 하나만 홀수인 경우
    elif is_a_odd or is_b_odd:
        return 2 * (a + b)
    
    # 3. 둘 다 짝수인 경우

````python
# 문제 10번 정답 확인 코드

def solution(a, b):
    # 기본 틀의 answer 변수를 활용합니다.
    answer = 0
    
    # a와 b가 홀수인지 미리 체크해두면 코드가 깔끔해집니다.
    a_is_odd = (a % 2 == 1)
    b_is_odd = (b % 2 == 1)
    
    # 1. 모두 홀수일 때
    if a_is_odd and b_is_odd:
        answer = a**2 + b**2
        
    # 2. 하나만 홀수일 때 (이미 위에서 둘 다 홀수인 경우는 걸러졌으므로 or만 쓰면 됩니다)
    elif a_is_odd or b_is_odd:
        answer = 2 * (a + b)
        
    # 3. 모두 짝수일 때
    else:
        answer = abs(a - b)
        
    return answer


# 테스트 케이스 실행
test_cases = [
    {"a": 3, "b": 5, "expected": 34},
    {"a": 6, "b": 1, "expected": 14},
    {"a": 2, "b": 4, "expected": 2}
]

for i, test in enumerate(test_cases):
    res = solution(test["a"], test["b"])
    print(f"입출력 예 #{i+1}")
    print(f"  - 입력: a={test['a']}, b={test['b']}")
    print(f"  - 결과: {res} (정답 여부: {'✅' if res == test['expected'] else '❌'})")
    print("-" * 30)
````
